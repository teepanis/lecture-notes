### <font color='darkcyan'> ข้อ 1) ทอดลูกเต๋า แล้ว $\text{ได้เงิน} = \text{(เลขหงาย)}^2$ โดยเฉลี่ยได้เงินกี่บาท?</font>

*เฉลย* $\quad$ อาศัยสูตรการหาค่าเฉลี่ย ด้วยความน่าจะเป็น

$$
\bar{x} = \sum_n P_n x_n \tag{1}
$$

ทุกหน้าที่หงายขึ้น มีความน่าจะเป็นเท่ากันคือ $P_n = \frac{1}{6}$ ดังนั้น

$$
\begin{aligned}
\bar{x}  & = \frac{1}{6}1^2 + \frac{1}{6}2^2 + \cdots \frac{1}{6}6^2 \\
 & = \frac{1}{6}[1+4+9+16+25+36]
\end{aligned}
$$

เราสามารถใช้ Python Sybmolic Package หาผลลัพธ์ข้างต้น ด้วย code ต่อไปนี้

```python
import sympy

terms = [1, 4, 9, 16, 25, 36]
total_sum = sum(terms)
result = sympy.Rational(1, 6) * total_sum

print(f"Expression: 1/6 * {terms}")
print(f"Symbolic Result: {result}","=",float(result))
```

ดังนั้น คำตอบคือ

$$
\boxed{ \bar{x} = \frac{91}{6} \approx 15.16 \text{ บาท} }
$$

### <font color='darkcyan'> ข้อ 2) จากนิยามของ Partition Function $Z = \sum_r e^{-E_r/k_B T}$ จงแสดงว่า  $\bar{E} = -\frac{\partial}{\partial \beta} \ln Z,\; \beta = k_B T$</font>

บอกใบ้: ให้เริ่มจาก Canonical Distribution

**เฉลย** Canonical Distribution บอกว่า ความน่าจะเป็น

$$
\begin{aligned}
P_r & \propto e^{-E_r/k_B T} = C e^{-E_r/k_B T}
\end{aligned}
$$

เมื่อ $C$ คือ ค่าคงที่ $\quad$ จากเงื่อนไข Normalization ของความน่าจะเป็น $\sum_r P_r = 1$ ดังนั้น

$$
\sum_r C e^{-E_r/k_B T} = 1 \; \rightarrow \; C = \frac{1}{\sum_r e^{-E_r/k_B T}} = \frac{1}{Z}
$$

เราจึงบอกได้ชัดเจนว่า ความน่าจะเป็นที่ระบบจะอยู่ในสถานะ $r$ ก็คือ

$$
P_r = \frac{1}{Z} e^{-E_r/k_B T}
$$

ใช้สูตรการคำนวณหา ค่าเฉลี่ย ในทำนอง สมการ (1) จะได้

$$
\bar{E} = \sum_r P_r E_r = \sum_r \frac{1}{Z} E_r e^{-E_r/k_B T} = \frac{1}{Z} \sum_r E_r e^{-E_r/k_B T}
$$

เทอมซัมเมชั่นสามารถปรับให้อยู่ในรูปของอนุพันธ์ $\frac{\partial}{\partial \beta}$ ดังนี้

$$
\sum_r E_r e^{-E_r/k_B T} = \sum_r E_r e^{-\beta E_r} = - \frac{\partial}{\partial \beta} \sum_r e^{-\beta E_r} = - \frac{\partial}{\partial \beta} Z
$$

กล่าวคือ อนุพันธ์ $\frac{\partial}{\partial \beta}$ จะดึงเอง $E_r$ ลงมาด้านล่าง พร้อมใส่เครื่องหมายลบ เพื่อให้เป็นบวกเหมืิอนเดิม &nbsp; เราจึงบอกได้ว่า

$$
\boxed{ \bar{E} = - \frac{1}{Z}\frac{\partial}{\partial \beta} Z = -\frac{\partial}{\partial \beta} \ln Z} \tag{2}
$$

### <font color='darkcyan'> ข้อ 3) จาก Entropy $S = k_B \ln \frac{(N+q-1)!}{q! (N-1)!}$ จงแสดงว่า $\frac{dS}{dq} \approx k_B \ln \frac{N+q}{q}$</font>

**เฉลย** เนื่องจากเทอม $\frac{(N+q-1)!}{q! (N-1)!}$ เป็นจำนวนเต็ม จึงหาอนุพันธ์ไม่ได้ เราต้องเปลี่ยนมันเป็นจำนวน *ต่อเนื่อง* &nbsp; โดยสามารถใช้ Stiring Approximation ดังนี้

$$
\ln(x!) \approx x \ln x - x
$$

จากเทอมในโจทย์จะได้ว่า

$$
\ln \frac{(N+q-1)!}{q! (N-1)!} = \ln (N+q-1)! - \ln q! - \ln (N-1)!
$$

ทำการใช้ Stirling Approximation กับทั้ง 3 เทอม:

$$
\ln \frac{(N+q-1)!}{q! (N-1)!} \approx (N+q-1)\ln (N+q-1) - (N+q-1) - q \ln q + q - (N-1)\ln (N-1) + (N-1)
$$

หากจัดการเทอมด้วยความระมัดระวัง สามารถหาอนุพันธ์เทียบ $q$ ของทางขวามือ หรือ เราใช้ Python Symbolic Package ได้ดังนี้

*หมายเหตุ* ผมป้อนคำสั่ง Google AI Mode ให้เขียนโค้ดให้ดังนี้: <font color='red'>
Show python symbolic package code that find derivative of (N+q-1)\ln (N+q-1) - (N+q-1) - q \ln q + q - (N-1)\ln (N-1) + (N-1) with respect to q</font>

```python
import sympy


N, q = sympy.symbols('N q')

expr = (N + q - 1) * sympy.log(N + q - 1) - (N + q - 1) - \
       q * sympy.log(q) + q - \
       (N - 1) * sympy.log(N - 1) + (N - 1)

derivative = sympy.diff(expr, q)
simplified_derivative = sympy.simplify(derivative)
print(f"Derivative: {simplified_derivative}")
```

จากผลการรันข้างต้น เราบอกได้ว่า

$$
\frac{dS}{dq}  \approx k_B \ln \frac{N+q-1}{q}
$$

แต่จากโจทย์จะเห็นว่า เราตัด (drop) (-1) ทิ้งไปได้ เพราะมันเป็นการประมาณที่สมเหตุผลเพราะ Stiring Approximation ใช้ในกรณีที่ $N, q$ มีค่าสูงมาก หากจะตัดออกสัก 1 ก็คงไม่เสียหายมากนั้น ดังนี้

$$
\boxed{\frac{dS}{dq}  \approx k_B \ln \frac{N+q}{q}}
$$

### <font color='darkcyan'> ข้อ 4) คุณคือผู้ทรงฯที่ Journal ส่งงานของ Einstein มาให้เช็คความถูกต้อง $\bar{E}_\text{QM} = \frac{\hbar \omega}{2} \coth(\frac{\hbar \omega}{ 2 k_B T})$ โดยที่พลังงานของ oscillator คือ $E_n = \hbar \omega (n+\frac{1}{2})$ &nbsp; จงหา $\bar{E}$ แล้วโปรดแนะนำทาง Journal ว่าควร ตีพิมพ์หรือไม่? [A.Einstein, Annalen der Physik, p.186 (1907)] (https://doi.org/10.1002/andp.19063270110)</font>

**เฉลย** เราจะใช้สูตรการหาค่าเฉลี่ยจาก สมการ (2) โดยกำหนดให้ระดับพลังงานของแต่ละสถานะ อยู่ในรูป

$$
E_r = \hbar \omega (r+\frac{1}{2})
$$

ดังนั้น

$$
\ln Z = \ln [ \sum_{r=0}^\infty e^{- \beta \, \hbar \omega (r+1/2)} ] = \ln [ e^{- \beta \hbar \omega /2} \sum_{r=0}^\infty x^r ], \; x \equiv e^{-\beta \hbar \omega } 
$$

เทอม $\sum \cdots$ คือ อนุกรมเรขาคณิต จึงได้เท่ากับ $\frac{1}{1-x}$ ดังนั้น

$$
\ln Z = \ln [ e^{- \beta \hbar \omega /2} \frac{1}{1-x} ] = -\beta  \frac{\hbar \omega}{2} + \ln \frac{1}{1-x} =  -\beta  \frac{\hbar \omega}{2} - \ln ( {1-e^{-\beta \hbar \omega}})
$$

จากนั้นหาอนุพันธ์เทียบ $\beta$ เพื่อคำนวณพลังงานเฉลี่ย


$$
\bar{E}_\text{QM} = - \frac{\partial }{\partial \beta} \ln Z = \frac{\hbar \omega}{2} + \frac{\hbar \omega e^{-\beta \hbar \omega}}{1-e^{-\beta \hbar \omega}} = \frac{\hbar \omega}{2}\frac{1+ e^{-\beta \hbar \omega}}{1- e^{-\beta \hbar \omega}} 
$$

คูณ $e^{\beta \hbar \omega/2}$ ทั้งบนและล่างของเศษส่วน จึงจัดให้อยู่ในรูปของ $\coth(\cdots)$ ได้ดังนี้

$$
\boxed{ \bar{E}_\text{QM} = \frac{\hbar \omega}{2} \coth(\frac{\hbar \omega}{2 k_B T}) }
$$

เราสามารถเทียบพลังงานนี้ กับกรณี Classical (Equipartition Theorem) ด้วยการวาดกราฟโดย Python ดังนี้


```python
import matplotlib.pyplot as plt
import numpy as np

hbarOmega = 2.
kBT = np.arange(0.1, 5, 0.1)
EQM = hbarOmega/2/np.tanh(hbarOmega/2/kBT)
EEQ = kBT

plt.plot(kBT,EQM,label='QM')
plt.plot(kBT,EEQ,label='Classical')
plt.grid()
plt.legend()
plt.xlabel('$k_B T$')
plt.ylabel('$\\bar{E}$')
plt.show()
```

จากกราฟจะเห็นว่าเมื่ออุณภูมิสูงขึ้น ทั้งกรณี Quantum และ Classical มีผลเดียวกันคือ พลังงานเฉลี่ยแปรตาม $k_B T$ ซึ่งเป็นผลเนื่องมาจาก equi-partition theorem &nbsp; ดังนั้น ในกรณีนี้ ในมุมของการทดลอง เราจะเห็นผลทางควอนตัมก็ต่อเมื่อ อุณหภูมิต่ำเท่านั้น 

ในรายงานวิจัยของ Einstein นั้น การเทียบกับการทดลอง โดยดูที่พลังงาน $\bar{E}$ นั้นเป็นไปได้ยาก เพราะ พลังงานมักเป็นปริมาณสัมพัท เช่น พลังงานศักย์โน้มถ่วง $U = m g h$ ก็ขึ้นอยู่กับว่าเราใช้จุดใดเป็นความสูง $h = 0$ &nbsp; Einstein ใช้การเปลี่ยนแปลงพลังงานเทียบอุณหภูมิ หรือ ตีความได้ว่า เป็น 

$$
\text{ความจุความร้อนจำเพาะ (คิดต่อโมล) } C_V =  \frac{d\bar{E}}{dT} 
$$

กรณี Classical โดยใช้ equi-partition theorem อะตอมอยู่ใน 3 มิติ ดังนั้น พลังงานของอะตอม 1 โมลคือ

$$
\bar{E}_\text{Classic} = 3 N_A k_B T \; \rightarrow \; C_V = 3 N_A k_B = 3 R \approx 24.9 \text{ J }\text{K}^{-1}\text{mol}^{-1}
$$

และกรณีควอนตัมคือ

$$
\bar{E}_\text{QM} = 3 N_A \frac{\hbar \omega}{2}\coth{\frac{\hbar \omega}{2 k_B T} } \; \rightarrow \; C_V = 3 R \left[ \frac{\hbar \omega}{2 k_B T \sinh(\frac{\hbar \omega}{2 k_B T})} \right]^2
$$

เราสามารถวาดกราฟเทียบกันดังนี้

```python
import matplotlib.pyplot as plt
import numpy as np

hbarOmega = 2.
kBT = np.arange(0.1, 2, 0.01)
CQM = (hbarOmega/2/kBT/np.sinh(hbarOmega/2/kBT))**2
CEQ = 0*kBT+1

plt.plot(kBT,CQM,label='QM')
plt.plot(kBT,CEQ,label='Classical')
plt.grid()
plt.legend()
plt.xlabel('$k_B T$')
plt.ylabel('$3 R$')
plt.show()
```

ข้างต้นเป็นกราฟเดียวกันที่ปรากฎในรายงานของ Einstein และถือเป็นก้าวแรกของการวิเคราะห์ ที่ใช้ควอนตัมมาประยุกต์กับ Solid State Physics สามาถอ่านเปเปอร์ได้จาก Wikipedia (https://en.wikipedia.org/wiki/List_of_scientific_publications_by_Albert_Einstein)
